```python
mini = fifa23['Wage'].min()
maxi = fifa23['Wage'].max()

ran = np.linspace(mini,maxi,100) #split range(min~max) into 100 sub-ranges

#number of data per each range
cnt = []
b_ran = 0
for i in ran[1:]:
  cnt.append(len(fifa23[(fifa23['Wage']<=i) & (fifa23['Wage']>b_ran)])) # <b_ran< wage <=i in ran[]
  b_ran = i

sort_cnt = sorted(cnt) #ascending order
sort_cnt_rev = list(reversed(sort_cnt)) #descending order

print(f'mid = {sort_cnt[len(sort_cnt)//2]}') #median
print(f'25= {sort_cnt[len(sort_cnt)//4]}') #Q1
print(f'75= {sort_cnt_rev[len(sort_cnt)//4]}')#Q3
print(f'mean = {sum(sort_cnt)/len(sort_cnt)}')#mean

mean_cnt = sum(sort_cnt)//len(sort_cnt)
mid_cnt=sort_cnt[len(sort_cnt)//2]
seven_cnt=sort_cnt_rev[len(sort_cnt)//4] #3Q
#to 각 구간의 개수를 맞추어 주기 위함
```

```python
select_data = pd.DataFrame()

b_ran = 0
ran[-1] += 1
for idx, val in enumerate(ran[1:]): #idx=i
  if cnt[idx] > seven_cnt: #if the number of data is larger then Q3, randomly select
     ran_data = fifa23[(fifa23['Wage']<val) & (fifa23['Wage']>=b_ran)].sample(n=seven_cnt,random_state=42)
  else: #if not, put all the data into ran_data
      ran_data = fifa23[(fifa23['Wage']<val) & (fifa23['Wage']>=b_ran)]

  print(len(ran_data))

  select_data = pd.concat([select_data,ran_data],ignore_index=True) #merging data frame, put select_data into ran_data

  print(f'select size = {len(select_data)}')

  b_ran = val

select_data
```
